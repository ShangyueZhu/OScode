```python
snmpdata_names = sorted(glob.glob('data-Stanford-09-29-18/*.snmp'))
```

```python
for y in snmpdata_names:
    TransFile = open(y)
    TransList = TransFile.readlines()
    half = int(len(TransList)/2)
    Transmitted = []
    Received = []
    APIP = []
    APName=[]
    totalUser = 0
    user5G = []
    for x in TransList[:half]:
        if "2.2.1.4.1.1.0" in x:
            totalUser = int(x.split(' ')[-1])
        elif "2.2.1.5.2.1.4.1.3." in x:
            APName.append(x.split(' ')[-1].split('"')[1])
        elif "2.2.1.5.2.1.4.1.2." in x:
            APIP.append((x.split(' ')[-1]).split('\n')[0])
        elif "2.2.1.4.1.2.1.26." in x:
            user5G.append(int(x.split(' ')[-1]))
        elif "5.3.1.1.1.21" in x:
            Transmitted.append(int(x.split(' ')[-1]))
        elif "5.3.1.1.1.19" in x:
            Received.append(int(x.split(' ')[-1]))
    timest = y.split('-')
    times = '-'.join(timest[5:8])+' '+timest[8]+':'+timest[9]+":"+timest[-1].split('.')[0]
    user5G = np.array(user5G)
    client5G = int(np.sum(user5G - 1)/3)
    insert_content = {"Timestamp": times, 
                      APName[0]: {"IP":APIP[0],"2_4G":{"Tx":Transmitted[0],"Rx":Received[0]}, 
                                        "5G":{"Tx":Transmitted[1],"Rx":Received[1]}},
                      APName[1]: {"IP": APIP[1],"2_4G":{"Tx":Transmitted[2],"Rx":Received[2]}, 
                                        "5G":{"Tx":Transmitted[3],"Rx":Received[3]}},
                      APName[2]: {"IP": APIP[2],"2_4G":{"Tx":Transmitted[4],"Rx":Received[4]}, 
                                        "5G":{"Tx":Transmitted[5],"Rx":Received[5]}},
                      APName[3]: {"IP": APIP[3],"2_4G":{"Tx":Transmitted[6],"Rx":Received[6]}, 
                                        "5G":{"Tx":Transmitted[7],"Rx":Received[7]}},
                      "Client": totalUser,"Client(5G)": client5G}
    print(insert_content)
    db.client_number.insert(insert_content)
```
