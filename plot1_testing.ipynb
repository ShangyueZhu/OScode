```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ReceFile = open("plot_r1.txt")
DestList = ReceFile.readlines()

TransFile = open("plot_t1.txt")
TransList = TransFile.readlines()
```

```python
AP1 = []
AP2 = []
AP3 = []
AP4 = []
increase = int((len(TransList)-2)/(int(TransList[1])))
item = int(TransList[1])


for x in range(2+increase+1,len(TransList),increase):
    gapAP1 = (int(TransList[x])/1024/1024) - (int(TransList[x-increase])/1024/1024)
    AP1.append(gapAP1)
    gapAP2 = (int(TransList[x+2])/1024/1024) - (int(TransList[x+2-increase])/1024/1024) 
    AP2.append(gapAP2)
    gapAP3 = (int(TransList[x+4])/1024/1024) - (int(TransList[x+4-increase])/1024/1024)
    AP3.append(gapAP3)
    gapAP4 = (int(TransList[x+6])/1024/1024) - (int(TransList[x+6-increase])/1024/1024) 
    AP4.append(gapAP4)

filetime = TransList[0].split('/')[-1].split('.')[0].split('-')
b = '-'.join(filetime[1:4])
c = ':'.join(filetime[4:5])    
xlabellist= []
for x in range(0,len(AP1),2):
        datatime = b+' '+c+':0%s'%x
        xlabellist.append(datatime)
AP1R = []
AP2R = []
AP3R = []
AP4R = []
for x in range(2+increase+1,len(DestList),increase):
    
    gapAPR1 = (int(DestList[x])/1024/1024) - (int(DestList[x-increase])/1024/1024)
    AP1R.append(gapAPR1)#ND-WI-275-1
    gapAPR2 = (int(DestList[x+2])/1024/1024) - (int(DestList[x+2-increase])/1024/1024)
    AP2R.append(gapAPR2)#ND-WI-275-2
    gapAPR3 = (int(DestList[x+4])/1024/1024) - (int(DestList[x+4-increase])/1024/1024)
    AP3R.append(gapAPR3) #ND-WI-275-4
    gapAPR4 = (int(DestList[x+6])/1024/1024) - (int(DestList[x+6-increase])/1024/1024)
    print(gapAPR4)
    AP4R.append(gapAPR4)#ND-WI-275-3
```

```python
%matplotlib notebook
x = np.arange(len(AP1))
plt.figure()
ax = plt.subplot(111)

ax.bar(x, AP1,width=0.1,color='b',align='center',label='AP1')
ax.bar(x+0.1, AP2,width=0.1,color='r',align='center',label='AP2')
ax.bar(x+0.2, AP4,width=0.1,align='center',label='AP3')
ax.bar(x+0.3, AP3,width=0.1,align='center',label='AP4')

ax.set_ylabel('Detla data transmitted (Mbytes)')
ax.set_title("(5G) Transmitted bytes in %s minutes"%(item-1))
ax.legend()
ax.set_xlabel("Time start on %s (minutes)"%xlabellist[0])

plt.grid(True)
# plt.savefig('/home/netscale/slackpost/Transmit')
plt.show()
```

```python
%matplotlib notebook
x = np.arange(len(AP1R))
plt.figure()
ax = plt.subplot(111)
# ax2 = ax.twinx()

ax.bar(x, AP1R,width=0.1,color='b',align='center',label='AP1')
ax.bar(x+0.1, AP2R,width=0.1,color='r',align='center',label='AP2')
ax.bar(x+0.2, AP4R,width=0.1,align='center',label='AP3')
ax.bar(x+0.3, AP3R,width=0.1,align='center',label='AP4')

ax.set_ylabel('Detla data Received (Mbytes)')
ax.set_title("(5G) Received bytes in %s minutes"%(item-1))
ax.legend()
ax.set_xlabel("Time start on %s (minutes)"%xlabellist[0])

plt.grid(True)
# plt.savefig('/home/netscale/slackpost/Received')
plt.show()
```
